```python
import requests
```

```python
import pandas as pd
```

#Step 1: Load SpaceX Launch Data from API


```python
launch_url="https://api.spacexdata.com/v4/launches"
response= requests.get(launch_url)
response
```

```python
data= response.json()
df_launch= pd.DataFrame(data)
launches =df_launch[['name', 'date_utc', 'success', 'details', 'rocket']]
launches
```

```python
launches['date_utc']=pd.to_datetime(launches['date_utc'])
launches['year']=launches['date_utc'].dt.year
launches
```

#Step 2: Load Rocket Metadata

```python
rocket_url=' https://api.spacexdata.com/v4/rockets'
response=requests.get(rocket_url)
response
```

```python
rocket_data=response.json()
df_rocket=pd.DataFrame(rocket_data)
rockets= df_rocket[['id', 'name', 'type', 'active', 'stages']]
rockets
```

#Step 3: Merge Launch and Rocket Data


```python
merger_df=pd.merge(launches,rockets, left_on='rocket', right_on='id')
merger_df=merger_df.drop(['rocket'],axis=1)
merger_df
```

#Step 4: Add Simulated Country Information


```python
import random
```

```python
countries= ['USA', 'Russia', 'India', 'China', 'France']
merger_df['country']=[random.choice (countries) for i in range(len(merger_df))]
merger_df
```

#Step 5: Store Merged Data in SQLite3


```python
import sqlite3
import pandas as pd
```

```python
conn= sqlite3.connect('launches.db')
merger_df.to_sql('launches',conn,if_exists='replace',index=False)
```

#Step 6: Run SQL Queries on the Data to analyze


```python
launches_by_country=pd.read_sql_query('SELECT country,COUNT(*)as count FROM launches GROUP BY country',conn)
launches_by_country
```

```python
year_high_no_of_launches= pd.read_sql_query('SELECT year,COUNT(*)as count FROM launches GROUP BY year ORDER BY count DESC LIMIT 1',conn)
year_high_no_of_launches
```

```python
top_5_mission_by_launch_count = pd.read_sql_query('SELECT name_x,COUNT(*)as count FROM launches GROUP BY name_x ORDER BY count DESC LIMIT 5',conn)
top_5_mission_by_launch_count
```
